```python
# Question 1
import pandas as pd
import numpy as np

from pandas import DataFrame

data=pd.read_csv('/Users/Behnoush/Documents/GitHub/Data-analyst-challenge/takeHome.csv')

data=pd.DataFrame(data, columns=['abt','ltv','ttime','tue'])

#data.dtypes
#data.describe()

data2=data.groupby('abt').mean()
data2=pd.DataFrame(data2, columns=['ltv','ttime','tue'])
data2.head()

n_groups = len(data2)
index = np.arange(n_groups)+1

import matplotlib as mpl
import matplotlib.pyplot as plt

w = 0.75

plt.figure(figsize=(12,8))
plt.bar(index, data2['ltv'], width=w, color='m', align='center',label='Life Time Value')

plt.xlabel('Group', fontsize=15)
plt.ylabel('Life Time Value', fontsize=15)
plt.title('This chart shows "Life time value" per AB group', fontsize=20)

plt.show()


plt.figure(figsize=(12,8))
plt.bar(index, data2['ttime'], width=w, color='y', align='center', label='Total time')

plt.xlabel('Group', fontsize=15)
plt.ylabel('Total Time', fontsize=15)
plt.title('This chart shows "Total time" per AB group', fontsize=20)

plt.show()


plt.figure(figsize=(12,8))
plt.bar(index, data2['tue'], width=w, color='c', align='center', label='Total Unites earned')

plt.xlabel('Group', fontsize=15)
plt.ylabel('Total Unites Earned', fontsize=15)
plt.title('This chart shows "Total Unites earned" per AB group', fontsize=20)

plt.show()
```
